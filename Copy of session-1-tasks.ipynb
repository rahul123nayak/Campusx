### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
num1 = int(input("enter 1st number"))
num2 = int(input("enter 2nd number"))
temp = None
print("first number is ",num1,"2nd number is ",num2)
temp = num1
num1 = num2
num2 = temp
print("after swap")
print("first number is",num1,"2nd number is ",num2)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
x1 = int(input("Enter 1st x cordinate"))
y1 = int(input("Enter 1st y cordinate"))
x2 = int(input("Enter 2nd X cordinate"))
y2 = int(input("Enter 2nd y cordinate"))

ecd_distance = (x2-x1)*2+(y2-y1)*2
print("euclidean distance between 2 co-ordinates are: ", ecd_distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
p = int(input("Enter principal amount"))
t = int(input("Enter time in months"))
r = int(input("Enter rate of interest"))
sim_interest = (p*t*r)/100
print("simple interest is: ",sim_interest)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
heads = int(input("Enter total number of heads"))
legs = int(input("Enter total number of legs"))
dogs = 0
chicken = 0

for i in range(0,heads,2):
  legs = legs-4
  dogs += 1
  legs = legs-2
  chicken += 1
print("number of dogs is :",dogs,"number of chickens is: ", chicken)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
n = int(input("enter the value of n"))
sum_square = (n*(n+1)*(2*n+1))/6
print("sum of squares of first ",n," natural number is: ",sum_square)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
a = int(input("enter the 1st term of AP"))
b = int(input("enter the 2nd term of AP"))
n = int(input("enter the nth term of AP"))
d = b-a
nth_term = a+(n-1)*d
print("nth term of series is ",nth_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python

x1 = int(input("Enter numerator of 1st fraction"))
y1 = int(input("Enter denominator of 1st fraction"))
x2 = int(input("Enter numerator of 2nd fraction"))
y2 = int(input("Enter denominator of 1st fraction"))

num = (x1*y2)+(x2*y1)
deno = y1*y2

print("sum of 2 fraction is: ",num,"/",deno)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math
ht = int(input("Enter height of tank"))
lt = int(input("Enter length of tank"))
bt = int(input("Enter breadth of tank"))
hg = int(input("Enter height of glass"))
rg = int(input("Enter radius of glass"))

volume_tank = ht*lt*bt
volume_glass = math.pi*rg**2*hg
print("number of glasses of milk obtained from tank is: ", math.ceil(volume_tank/volume_glass))
````
